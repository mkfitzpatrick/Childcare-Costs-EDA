Data Source

Handling Missing Values: Columns with missing values, such as _75FCCInfant_flag and related fields, were excluded to retain consistency in the analysis. This removed 11,184 rows with incomplete data.

Selecting Pertinent Columns: From the extensive dataset, 29 relevant columns were chosen to focus on variables directly tied to the project goals, such as childcare costs, income, and economic indicators.

Final Dataset: After cleaning, the dataset contained 23,342 rows and 29 columns, providing a concise subset for analysis.

## Results

```python
import pandas as pd

df = pd.read_excel('cleaned_data.xlsx')


df.describe()
```

1. Normalization of Variables
I normalized certain variables (e.g., Unemployment Rate and Female Labor Force Participation Rate) to compare them on a consistent scale, especially given their varying magnitudes.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()


df[['UNR_16', 'FLFPR_20to64']] = scaler.fit_transform(df[['UNR_16', 'FLFPR_20to64']])


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='StudyYear', y='UNR_16', label='Normalized UNR_16', marker='o', color='blue')
sns.lineplot(data=df, x='StudyYear', y='FLFPR_20to64', label='Normalized Female Labor Force Participation', marker='x', color='red')

plt.title('Normalized Unemployment Rate and Female Labor Force Participation Rate Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Normalized Value', fontsize=12)
plt.legend(title='Variables')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Findings:
Normalizing these variables allows us to visualize trends over time without the influence of differing units or scales.
We observe that the Unemployment Rate (UNR_16) increases in 2013, while the Female Labor Force Participation Rate (FLFPR_20to64) declines at the same time, suggesting a possible link between these two factors during that period.


2. Correlation Analysis
I conducted a correlation analysis to examine relationships between key socioeconomic factors such as income, unemployment, and childcare costs.

```python

correlation_matrix = df[['MHI', 'MCPreschool', 'PR_F', 'PR_P', 'FLFPR_20to64', 'MLFPR_20to64', 'UNR_16', 'FUNR_16', 'ME', 'FME', 'MME']].corr()


plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Between Key Variables')
plt.show()
```

Findings:
Income and Childcare Costs: There is a moderate positive correlation between median household income (MHI) and the cost of preschool childcare (MCPreschool). This suggests that higher income areas may face higher childcare costs.
Labor Force Participation and Unemployment: Unemployment (UNR_16) is negatively correlated with female labor force participation (FLFPR_20to64), which indicates that higher unemployment may discourage women from participating in the labor force.

Childcare Costs and Poverty Rates
We investigate the relationship between childcare costs and poverty rates, specifically focusing on the Family Poverty Rate (PR_F) and its impact on childcare costs.

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PR_F', y='MCPreschool', hue='StudyYear', palette='coolwarm', marker='o')

plt.title('Childcare Costs vs. Family Poverty Rate', fontsize=16)
plt.xlabel('Poverty Rate for Families', fontsize=12)
plt.ylabel('Cost of Childcare (Preschool)', fontsize=12)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Findings:
Negative Correlation: A negative relationship between poverty rates (PR_F) and the cost of childcare suggests that regions with higher poverty rates tend to have lower childcare costs. This could be because of a lower demand for childcare in economically depressed areas or a greater availability of subsidies.
This finding supports the hypothesis that childcare costs are more affordable in areas with greater financial need, though the affordability challenge persists due to income disparities.

Childcare Costs and Female Labor Force Participation
We explore the link between childcare costs and female labor force participation (FLFPR_20to64).

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='FLFPR_20to64', y='MCPreschool', hue='StudyYear', palette='plasma', marker='o')

plt.title('Childcare Costs vs. Female Labor Force Participation', fontsize=16)
plt.xlabel('Female Labor Force Participation Rate (20 to 64)', fontsize=12)
plt.ylabel('Cost of Childcare (Preschool)', fontsize=12)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Findings:
Moderate Correlation: There is a moderate correlation between female labor force participation and childcare costs. This suggests that as female labor force participation increases, the demand for childcare services increases, which can drive up costs.
However, there are clusters of data where high childcare costs are associated with lower labor force participation, which may indicate that high costs may discourage some women from entering or remaining in the workforce

5. Percent Change Analysis

To better understand the rate at which both childcare costs and median household income have changed over time, I calculated the percent change for both variables year-over-year. This provides a clearer picture of how much each variable has grown or declined relative to the previous year.

```python

df['MHI_pct_change'] = df['MHI'].pct_change() * 100
df['MCPreschool_pct_change'] = df['MCPreschool'].pct_change() * 100


print(df[['StudyYear', 'MHI_pct_change', 'MCPreschool_pct_change']])
```

Findings:
In 2014, childcare costs (MCPreschool) surged by 37.26%, while median household income (MHI) declined by 2.77%, highlighting a critical period where childcare became more expensive despite stagnant or falling incomes. Over time, while MHI fluctuated, childcare costs consistently increased, pointing to a growing gap between income growth and childcare affordability. The percent change analysis clearly illustrates these shifts and emphasizes the economic pressure families may face as childcare costs outpace income growth.
